```python
from lib import huffman
```

```python
probabilities = huffman.get_probabilities("hello world")
probabilities
```

```python
tree = huffman.build_tree(probabilities)
huffman.to_graphviz(tree)
```

```python
codes = huffman.get_codes(probabilities, tree)
codes
```

```python
encoded = huffman.encode("hello world", codes)
encoded
```

```python
decoded = huffman.decode(encoded, codes)
decoded
```
