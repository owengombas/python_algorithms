```python
from lib.tree import (
    Node,
    BinaryTreePostOrderIterator,
    BinaryTreeInOrderIterator,
    BinaryTreePreOrderIterator,
    BinaryTreeLevelOrderIterator,
)
```

```python
root = Node[str]("Root")
a, b = Node[str]("A"), Node[str]("B")
c, d = Node[str]("C"), Node[str]("D")
e, f = Node[str]("E"), Node[str]("F")

root.add_children(a, b)
b.add_children(c, d)
a.add_children(e, f)

root.to_graphviz()
```
